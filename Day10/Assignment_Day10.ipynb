```python
from scipy.fft import fft, fftfreq
```

```python
Fs = 500  
t = np.linspace(0, 1, Fs, endpoint=False)  
signal = np.sin(2 * np.pi * 5 * t) + np.sin(2 * np.pi * 50 * t)


fft_values = fft(signal)
freqs = fftfreq(Fs, 1 / Fs)

fig, axs = plt.subplots(2, 1, figsize=(10, 8))


axs[0].plot(t, signal, label='Signal (5Hz + 50Hz)')
axs[0].set_title('Time Domain Signal (5Hz + 50Hz)')
axs[0].set_xlabel('Time [s]')
axs[0].set_ylabel('Amplitude')
axs[0].grid(True)

axs[1].plot(freqs[:Fs//2], np.abs(fft_values[:Fs//2]))
axs[1].set_title('FFT of 5Hz + 50Hz Signal')
axs[1].set_xlabel('Frequency [Hz]')
axs[1].set_ylabel('Magnitude')
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

## E. Plotting normal random numbers and show that addition of two Gaussian is Gaussian and addition of all is Gaussian

```python
np.random.seed(0)
```

```python
x1 = np.random.normal(0, 1, 1000)
x2 = np.random.normal(0, 1, 1000)
x3 = np.random.normal(0, 1, 1000)
x4 = np.random.normal(0, 1, 1000)
x5 = np.random.normal(0, 1, 1000)
x6 = np.random.normal(0, 1, 1000)
x7 = np.random.normal(0, 1, 1000)
```

```python

sum_two = x1 + x2
sum_all = x1 + x2 + x3 + x4 + x5 + x6 + x7
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(x1, bins=30, alpha=0.5, label='x1')
plt.hist(x2, bins=30, alpha=0.5, label='x2')
plt.hist(sum_two, bins=30, alpha=0.7, label='x1 + x2', color='magenta')
plt.legend()
plt.title('Sum of Two Gaussian Distributions')
```

```python

plt.subplot(1, 2, 2)
plt.hist(sum_all, bins=30, alpha=0.7, label='Sum of All Gaussian', color='green')
plt.legend()
plt.title('Sum of Many Gaussian Distributions')
plt.tight_layout()
plt.show()
```
